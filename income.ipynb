```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#CSV file
income_florida = "income_florida.csv"
income_disparity = pd.read_csv(income_florida)

income_disparity
```

```python
#Graph for Graduation Rates

labels = ["Suncoast", "Southern", "Southeast", "Northwest", "Northeast", "Central"]

sizes = [76.0, 76.6, 76.2, 80.2, 68.0, 77.3]

plt.bar(labels, sizes)
plt.xlabel('Regions', fontsize=10)
plt.ylabel('Percentage', fontsize=10)
plt.title("Graduation Rate Avg by Region (2008-2018)")

plt.savefig("Images/'graduation'")
```

```python
#Graph for Unemployed Rates

labels = ["Suncoast", "Southern", "Southeast", "Northwest", "Northeast", "Central"]

sizes = [2.8, 3.15, 3.23, 4.08, 3.28, 3.58]

plt.bar(labels, sizes)
plt.xlabel('Regions', fontsize=10)
plt.ylabel('Unemployment %', fontsize=10)
plt.title("Unemployment Rate Avg by Region (2008-2018)")

plt.savefig("Images/'graduation'")
plt.savefig("Images/'unemployment'")
```

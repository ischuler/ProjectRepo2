```python
# Placement Moves per 1,000 Bed Days in Foster Care 
# Measure:The rate at which children change placements while in Foster care
# The purpose of this chart it to compare the placement moves % by regions 2009-2019Q2
```

```python
%matplotlib notebook
```

```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
placement_moves = pd.read_csv("Resources/pmregions.csv")

placement_moves
```

```python
# Filter the datframe for colums Entity & Total
piechart = placement_moves[["Entity","RegionAvg"]]

piechart
```

```python
# Labels for the sections of our pie chart
labels = ["Central", "Northeast", "Northwest", "Southeast", "Southern","Suncoast"]

# The values of each section of the pie chart
sizes = [9, 10, 12, 15, 21,33]

# The colors of each section of the pie chart
colors = ["orange", "lightblue", "purple", "green", "gray", "red"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0, 0, 0, 0,0,.1)
```

```python
# Creates the pie chart based upon the values above
# Automatically finds the percentages of each part of the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=145)

plt.title("2009-2019 Foster Care Placement Moves Per 1,000 Bed Days By Region")
```

```python
plt.tight_layout()
plt.savefig("Images/PlacementMovesPieChartByRegion")
plt.show()
```

```python
# Placement Moves per 1,000 Bed Days in Foster Care 
# Measure:The rate at which children change placements while in Foster care
# The purpose of this chart it to compare the placement moves % male vs.female 2009-2019Q2 
```

```python
# Read CSV
placement_moves = pd.read_csv("Resources/pmmalevfemale.csv")

placement_moves
```

```python
# Filter the placement_moves for colums "Entity","FemalePercentage","MalePercentage"
bargraph = placement_moves[["Entity","FemalePercentage","MalePercentage"]]

bargraph.head()
```

```python
bargraph.set_index('Entity').plot.bar(rot=0, title='2009-2019 Foster Care Placement Moves Per 1,000 Bed Days By Gender',
                                      figsize=(10,5), fontsize=12)
plt.xlabel("Florida Regions")
plt.ylabel("% of Placement Moves")
```

```python
plt.tight_layout()
plt.savefig("Images/PlacementMovesBarChartByGender")
plt.show()
```
